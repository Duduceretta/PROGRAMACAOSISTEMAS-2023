```python
def conversaoValorParaBinario(valor):

  if valor == 0 : return 0

  valorBinario = ""
  valorEhFloat = isinstance(valor, float)
  valorAbsoluto = abs(valor)

  while(valorAbsoluto >= 1):
    valorResto = int(valorAbsoluto % 2)
    valorAbsoluto /= 2
    valorBinario += str(valorResto)

  #adicionando zeros ao valor binario até que chegue em 8 bits
  while len(valorBinario) != 8:
    valorBinario += "0"

  valorInvertido = valorBinario[::-1]

  if(valorEhFloat):

    #separando a parte decimal da parte inteira
    stringValor = str(valor)
    parteDecimalString = "0." + stringValor.split(".")[1]
    parteDecimalFloat = float(parteDecimalString)
    parteDecimalBinaria = ""
    valorMultiplicado =  0

    #multiplicando a parteDecimalFloat por 2 sucessivamente e guardando resultado e multiplicando-o por 2
    for i in range(5):
      resultado = parteDecimalFloat * 2 if i == 0 else valorMultiplicado * 2
      resultadoString = str(resultado)
      parteDecimalBinaria += resultadoString.split(".")[0]
      parteDecimalResultado = float("0." + resultadoString.split(".")[1])
      valorMultiplicado = parteDecimalResultado

    valorInvertido += parteDecimalBinaria


  if(valor < 0):
    complementoDeDois = ''.join(map(lambda x : '0' if x == '1' else '1', valorInvertido))

    #adicionando 1 ao complemento de dois para obter o valor negativo em binario
    if(complementoDeDois[-1] == '0'):
      complementoDeDois = complementoDeDois[:-1] + '1'
      #inserindo '.' apos os 8 bits da parte inteira do numero se este for fracionario
      return complementoDeDois[:8] + "." + complementoDeDois[8:] if valorEhFloat else complementoDeDois

    else:
      valorConvertido = ""
      complementoDeDoisReverso = complementoDeDois[::-1]

      #iterando sobre o complemento de dois reverso para poder somar +1 ao mesmo, pois soma deve começar da direita para a esquerda
      for i in range(len(complementoDeDoisReverso)):

        if(complementoDeDoisReverso[i] == '1'):
          valorConvertido += '0'
        else:
          valorConvertido += '1' + complementoDeDoisReverso[i + 1:]
          break

      valorConvertidoReverso = valorConvertido[::-1]

      #inserindo '.' apos os 8 bits da parte inteira do numero se este for fracionario
      return valorConvertidoReverso[:8] + "." + valorConvertidoReverso[8:] if valorEhFloat else valorConvertidoReverso

  return valorInvertido[:8] + "." + valorInvertido[8:] if valorEhFloat else valorInvertido


def NumeroEhFloat(string):
    try:
      float(string)
      return True
    except ValueError:
      return False

def NumeroEhInt(string):
  try:
    int(string)
    return True
  except ValueError:
    return False

while True:

  numero = input("Digite um numero para converter para binario: ")
  if(NumeroEhInt(numero)):
    print(f"Valor de {numero} em binario: {conversaoValorParaBinario(int(numero))}")
    break
  elif(NumeroEhFloat(numero)):
    print(f"Valor de {numero} em binario: {conversaoValorParaBinario(float(numero))}")
    break
  else:
    print()
    print("Voce nao digitou um numero, tente de novo!")

```
